# matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in a jupyter
%matplotlib inline
import matplotlib.pyplot as plt
#Prepare some data
x=[1,2,3,4]
y=[11,222,33,44]
# 2.Setup plot
fig,ax=plt.subplots(figsize=(10,10)) # height and width
# 3.Plot the data
ax.plot(x,y)
#4.customize plot
ax.set(title="Simple plot",xlabel="X-axis", ylabel="Y-axis")
#5.Save & show (Save the whole figure)

fig.savefig("Sample Plot")
```

```python
#let's  try on another dataset
heart_disease=pd.read_csv("C:\Sample_Project1\zero-to-mastery-ml-master\data\heart-disease.csv")
heart_disease
```

```python
#Create a histogram
heart_disease["age"].plot.hist(bins=30)
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50
```

```python
len(over_50)
```

```python
#pyplot method
over_50.plot(x="age",y="chol",kind="scatter",c="target")
```

```python
fig,ax=plt.subplots(figsize=(10,5))
over_50.plot(x="age",y="chol",kind="scatter",c="target",ax=ax)
#ax.set_xlim([45,100])
```

```python
# OO method from scratch
fig,ax=plt.subplots(figsize=(10,5))
scatter=ax.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"])
# Customize the plot
ax.set(title="Heart diseases and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol "
       )
ax.legend(*scatter.legend_elements(),title="Target")
# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
# Subplot of age, chol, thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True
                           )
# Add data to ax0
scatter=ax0.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"])
# Customize ax0
ax0.set(title="Heart diseases and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol "
       )
# Add legend
ax0.legend(*scatter.legend_elements(),title="Target")
# Add meanline
ax0.axhline(over_50["chol"].mean(), linestyle="--")

#Add data to ax1
scatter=ax1.scatter(x=over_50["age"], y=over_50["thalach"],c=over_50["target"])
# Customize ax1
ax1.set(title="Heart diseases and Max heart rate",
       xlabel="Age",
       ylabel="Max heart rate"
       )
# Add legend
ax1.legend(*scatter.legend_elements(),title="Target")
# Add meanline
ax1.axhline(over_50["thalach"].mean(), linestyle="--")
fig.suptitle("Heart disease anaiysis", fontsize=16, fontweight="bold")
```
